```python
url="https://stackoverflow.com"
path="C:/Users/sanje/Desktop/jubi-solutions/chromedriver"
driver = webdriver.Chrome(path)
driver.get(url)
driver.maximize_window()
search=driver.find_element_by_name('q')
search.send_keys('node.js' + Keys.RETURN)

#Filtering on the basis of More -> Votes
driver.find_elements_by_xpath('//*[@id="mainbar"]/div[4]/div/div[2]/div/div[1]/button')[0].click()
driver.find_elements_by_xpath('//*[@id="uql-more-popover"]/ul/li[4]/a')[0].click()

#Clicking on 50 results per page
element = driver.find_element_by_xpath('//*[@id="mainbar"]/div[6]/a[3]')
driver.execute_script("arguments[0].scrollIntoView();", element)
element.click()

#Comments stores all the data
comments = pd.DataFrame(columns = ['id','url','header','votes','message',['votes','answer']]) 
```

```python
while (driver.find_elements_by_xpath('//*[@id="mainbar"]/div[7]/a[6]')[0]):

    #For one page iteration
    ids = driver.find_elements_by_xpath("//*[contains(@id,'question-summary-')]")
    comment_ids = []
    
    for i in ids:
        comment_ids.append(i.get_attribute('id'))
    print(comment_ids)

    for x in comment_ids:
        print("================================================================================================================")

        #Scrolling the elements into view
        element = driver.find_element_by_id(x)
        driver.execute_script("arguments[0].scrollIntoView();", element)

        #Clicking each comments
        driver.find_elements_by_xpath('//*[@id="'+x+'"]/div[2]/h3/a')[0].click()
        print('Question ID ::::::::',x)

        #Getting the url for reference
        url = driver.current_url
        print('Question URL :::::::',url)

        #Extracting the heads
        header = driver.find_elements_by_xpath('//*[@id="question-header"]/h1/a')[0].text
        print('Question Header ::::',header)

        #Extracting the message
        message = driver.find_elements_by_xpath('//*//*[@id="question"]/div[2]/div[2]/div[1]')[0].text
        print('Question :::::::::::',message)

        #Extracting the votes count
        votes = driver.find_elements_by_xpath('//*[@id="question"]/div[2]/div[1]/div/div')[0].text
        print('Question Votes :::::',votes)

        answers = pd.DataFrame(columns = ['votes','answer'])
        
        #Getting inside Answers
        ans_ids = driver.find_elements_by_xpath("//*[contains(@id,'answer-')]")
        answer_ids = []
        for a in ans_ids:
            answer_ids.append(a.get_attribute('id'))
        print(answer_ids)
        
        #Extracting all answers relating to a question
        for a_id in answer_ids:
            print("================================================================================================================")

            #Scrolling the elements into view
            element = driver.find_element_by_id(a_id)
            driver.execute_script("arguments[0].scrollIntoView();", element)

            #Extracting the answer votes count
            ans_votes = driver.find_elements_by_xpath('//*[@id="'+a_id+'"]/div/div[1]/div/div[1]')[0].text
            print('Answer Votes :::::::',ans_votes)

            #Extracting the answer message
            answer = driver.find_elements_by_xpath('//*[@id="'+a_id+'"]/div/div[2]/div[1]')[0].text
            print('Question :::::::::::',answer)

            #Saving votes and answer in answers
            answers.loc[len(answers)] = [ans_votes,answer]

        #Adding the data extracted to our main data-storage
        comments.loc[len(comments)] = [x,url,header,votes,message,answers]

        #Going back to previous page for next iteration
        driver.back()
    
    #Going to the next page
    driver.find_elements_by_xpath('//*[@id="mainbar"]/div[7]/a[6]')[0].click()
    
    
display(comments)
```
